```python
import boto3
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

s3 = boto3.resource('s3')
bucket = s3.Bucket('danhtran358-data-445-bucket')
fileObject = bucket.Object('diabetes.csv')
fileContentStream = fileObject.get().get('Body')

diabetes = pd.read_csv(fileContentStream)
diabetes
```

```python
## Preprocessing - Clean missing data values
## Glucose missing values
diabetes_cleaned = diabetes.loc[diabetes['Glucose'] != 0]
diabetes_cleaned
```

```python
## BloodPressure missing values
diabetes_cleaned = diabetes_cleaned.loc[diabetes_cleaned['BloodPressure'] != 0]
diabetes_cleaned
```

```python
## SkinThickness missing values
diabetes_cleaned = diabetes_cleaned.loc[diabetes_cleaned['SkinThickness'] != 0]
diabetes_cleaned
```

```python
## Insulin missing values
diabetes_cleaned = diabetes_cleaned.loc[diabetes_cleaned['Insulin'] != 0]
diabetes_cleaned
```

```python
## BMI missing values
diabetes_cleaned = diabetes_cleaned.loc[diabetes_cleaned['BMI'] != 0]
diabetes_cleaned
```

```python
## function to write write data_frame to csv file object in S3 bucket
def write_data_to_s3(file_name, data_frame):
    ## file object in s3 bucket
    data_file = bucket.Object(file_name)
    
    ## add content from the lists of recall scores
    content = data_frame.to_csv(index=False)

    ## store as new csv file
    data_file.put(Body = content)
```

```python
## write cleaned data to s3 bucket
write_data_to_s3('project_cleaned_data.csv', diabetes_cleaned)
```
